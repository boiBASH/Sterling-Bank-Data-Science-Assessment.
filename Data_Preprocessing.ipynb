```python
#Checking for missing values
missing = df.isnull().sum().to_frame('missing_count')
missing
```

```python
# Missing-value summary
missing_summary = df.isnull().sum().sort_values(ascending=False)
missing_percent = (missing_summary / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_summary,
    'Missing %': missing_percent
})
missing_df = missing_df[missing_df['Missing Count'] > 0]
missing_df
```

# **Data Cleaning Highlighted In Steps**

```python
# Step 1: Drop 'marital_status' due to very high missing rate
df.drop(columns=['marital_status'], inplace=True)

# Step 2: Replace missing 'employment_status' with 'Unknown'
df['employment_status'].fillna('Unknown', inplace=True)

# Step 3: Calculate age from 'date_of_birth' and drop the original column
today = pd.to_datetime("today")
df['age'] = df['date_of_birth'].apply(lambda x: (today - x).days // 365 if pd.notnull(x) else None)
df.drop(columns=['date_of_birth'], inplace=True)

# Step 4: Impute 'DAY_SINCEPAYMENT' with median
median_payment_days = df['DAY_SINCEPAYMENT'].median()
df['DAY_SINCEPAYMENT'].fillna(median_payment_days, inplace=True)

# Step 5: Fill missing 'DIRECTORATENAME' with 'Unknown'
df['DIRECTORATENAME'].fillna('Unknown', inplace=True)

# Step 6: Drop identifier and redundant columns
columns_to_drop = [
    'loan_id',          # identifier
    'industry_x',       # redundant with industry_desc
    'industry_y',       # numeric code for industry
    'sector_desc',      # duplicate of 'sector'
    'industry_desc'     # keep only if highly detailed vs 'LINE_DESC'
]
df.drop(columns=columns_to_drop, inplace=True)

# Step 7: Feature Engineering from dates
today = pd.to_datetime("today")
df['loan_age_days'] = (today - df['Contract_eff_date']).dt.days
df['customer_tenure_days'] = (today - df['customer_inception_date']).dt.days

# Step 8: Drop raw date columns used in feature engineering
df.drop(columns=['Contract_eff_date', 'customer_inception_date'], inplace=True)

# Preview the cleaned dataset before EDA
df_ready_preview = df.head()
df_ready_preview
```
